```python
import requests
import pandas as pd

# Replace 'YOUR_API_KEY' with your actual API key
api_key = 'f1ab339ca610b93824971611ebe7fe27'
sport = 'basketball'
region = 'us'  # Region: 'us', 'uk', 'eu', 'au'
market = 'h2h'  # Market: 'h2h', 'spreads', 'totals'

url = f'https://api.the-odds-api.com/v4/sports/{sport}/odds'
params = {
    'apiKey': api_key,
    'regions': region,
    'markets': market,
    'oddsFormat': 'decimal',
    'dateFormat': 'iso'
}

response = requests.get(url, params=params)

if response.status_code == 200:
    odds_data = response.json()
    
    # Prepare data to create a DataFrame
    all_data = []
    for event in odds_data:
        home_team = event.get('home_team')
        away_team = event.get('away_team')
        commence_time = event.get('commence_time')
        
        for bookmaker in event['bookmakers']:
            bookmaker_name = bookmaker.get('title')
            
            for market in bookmaker['markets']:
                for outcome in market['outcomes']:
                    all_data.append({
                        'Date': commence_time,
                        'Bookmaker': bookmaker_name,
                        'Home Team': home_team,
                        'Away Team': away_team,
                        'Outcome': outcome.get('name'),
                        'Odds': outcome.get('price')
                    })

    # Create DataFrame
    df_odds = pd.DataFrame(all_data)

    # Display or save the DataFrame
    print(df_odds)
    # Save to CSV if needed
    df_odds.to_csv('basketball_odds_data.csv', index=False)

else:
    print(f'Failed to retrieve data: {response.status_code}, Reason: {response.text}')
```
